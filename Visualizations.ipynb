```python
import pandas as pd #importing pandas library
import numpy as np #importing numpy library
import matplotlib.pyplot as plt #importing matplotlib library
import seaborn as sns #importing seaborn library
df2=pd.read_csv('final data set4.csv',keep_default_na=False, #reading the dataset  with specific types of missing values
                 na_values=['', 'NA', 'null','NaN'])
df2.head() #viewing the beginning observations of data set
```

```python

plt.figure(figsize=(16,9))
sns.relplot(x='Stress Score',y='Risk Index',kind='scatter',data=df2,hue='Exercise Days',size='Exercise Days',palette='Blues')#creating a scatter plot by using relplot function seperating hue by Exercise Days and adjusting size of the points by Exercise Days and setting the color palette as "Blues"
plt.title('Risk Index vs Stress Score',y=0.95)#setting the title as Risk Index vs Stress Score
plt.savefig('scatter2.png') #saving the visualization as png format
plt.show() #displaying the visualization
```

```python
plt.figure(figsize=(20,10)) #adjusting figure size
sns.catplot(x='Family History',y='Risk Index',data=df2,kind='box',hue='Drug Type',palette='pastel',showfliers=False) #creating a comparative box plot by using catplot() function seperating hue by Drug Type and hiding the outliers by using showfliers parameter and setting the palette as Blues
plt.title('Differing Values of Risk Index by Drug Type and Family History',y=0.98) #setting a title for the comparative box plot
plt.savefig('Comparative Box Plot01.png') #saving the visualization
plt.show() #displaying the visualization
```

```python
palette1 = {"Male": "#2986cc", "Female": "#c90076"}
#creating a custom palette

sns.set_style("whitegrid")
#setting the style



plt.figure(figsize=(16, 9))
#changing the figure size



sns.violinplot(data=df2, x="Drug Type", y="Risk Index", hue="Gender", split= True, inner="quartile", linewidth=1.25,palette=palette1)
#changing the line size and setting the custom palette


plt.title("Distribution of Risk Index by Gender and Drug Type", fontsize=20, y=1.05)
plt.xlabel("Drug Type", fontsize=15)
plt.ylabel("Risk Index", fontsize=15)
#changing title position, and the font sizes

plt.legend(title="Gender", loc="upper right")
#adding legend




plt.savefig("risk_index_violin.png")
#saving the figure


plt.show()
```

```python
sns.set_style("whitegrid")
#Setting up a grid so it is easier to gauge the values of different intervals

bins = [-1, 0.100, 0.200, 0.300, 0.400, 0.500, 0.600,0.700,0.800]
labels = ["<0.100", "0.1010.200", "0.2010.300", "0.3010.400", "0.401-0.500", "0.501-0.600","0.601-0.700","0.700<"]
#Grouping up the data as bins with labels to make it easier to visualize

df2["Risk Indexes Grouped"] = pd.cut(df2["Risk Index"], bins=bins, labels=labels)
#Setting up the bins as another column in the dataframe to make it usable in the visual

sns.catplot(kind="count",data=df2,y="Risk Indexes Grouped",col="Drug Awareness Program",hue="Drug Awareness Program",col_order=["None","Some","Extensive"],order=list(reversed(labels)))
#creating the visuals with three columns separating how often did the subjects attend the drug awareness program

plt.show()#displaying the visualization

plt.savefig("drug_awareness_program_effectiveness.png")
#Saving the final visual
```
