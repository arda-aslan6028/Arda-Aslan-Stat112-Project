```python
import pandas as pd  # importing libraries
import numpy as np


df=pd.read_csv('usa_drugs_risk_dataset_dirty.csv') # reading the csv file in google colab
df.head() # observing the first 5 rows to detect intial mistakes
```

```python
df.shape # getting the row number and column number (1020 rows, 16 columns)
```

```python
df.isna().sum()
```

```python
df.dtypes # Getting the data types in order to detect whether there are mistake in data type
```

```python
#fixing typos in column names and removing special characters in column names and making in title format for all column names
df=df.rename(columns={"STATE":"State","Age(yrs)":"Age","gEnder":"Gender","employment-status":"Employment Status","Income$":"Income","mentalHealthScore":"Mental Health Score","drug_useFrequency":"Drug Use Frequency","PeerInfluence":"Peer Influence","Family_History":"Family History","exerciseDays":"Exercise Days","DrugAwarenessProgram":"Drug Awareness Program","Risk_Index":"Risk Index"})
df.head()
```

```python
df.columns #checking column names in order to be sure about whether there are any mistakes
```

```python
df.columns = df.columns.str.strip() # an additional fix for Peer Influence since the first version of it does not contain a space between two words
df=df.rename(columns={"PeerInfluence":"Peer Influence"}) #renaming the peer influence column again
df.columns #observing the column names
```

```python
df[df.duplicated()] # observing the duplicate values
```

```python
df=df.drop_duplicates() #removing the duplicate values and later on we checked the shape( row number , column number) again (1011 rows 16 columns)
df.shape
```

```python

df['State']=df['State'].str.strip() #removing the leading and trailing space in values of State column
df['State']=df['State'].str.replace('#','',regex=False) #removing # character in values of State columns (regex=False for python to see the character as normal string)
df['State']=df['State'].str.replace('*','',regex=False) #removing * character in values of State columns (regex=False for python to see the character as normal string)
df['State']=df['State'].str.replace('$','',regex=False) #removing $ character in values of State columns (regex=False for python to see the character as normal string)
df['State']=df['State'].str.title() #making all values in State column in title format (arizona-->Arizona)
df['State']=df['State'].fillna(df['State'].mode()[0]) #filling na values of State column with mode of State column (there can be multiple modes so we take the first element of the modes)
df['State'].unique() #check the values whether they are unique after fixing


```

```python
df['Year']=df['Year'].fillna(df['Year'].mode()[0]) #filling missing values in year column with the first element of the modes of the year column
df['Year']=df['Year'].astype('Int64') #converting the data type of year column into Date --> 'Int64'
df['Year'] #viewing the new year column
```

```python
df['Age']=df['Age'].fillna(df['Age'].median()) #filling missing values in age column with the median of the age column
df['Age']=df['Age'].astype(int) # converting the data type of the age column into integer
df['Age'] #viewing the new age column
```

```python
df['Gender']=df['Gender'].str.title() #making every entry in gender column as title format
df['Gender']=df['Gender'].str.strip() #eliminating the leading and trailing spaces of entries if there are any.
df['Gender']=df['Gender'].str.replace('#','',regex=False) #removing the '#' character in the entries if there are any
df['Gender']=df['Gender'].str.replace('*','',regex=False) #removing the '*' character in the entries if there are any
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])  #filling missing values of gender column with the first element of the modes of gender column
df['Gender'].unique() #checking the gender types whether there are any mistake.

```

```python
df['Education Level'].value_counts() #checking the counts for education level types so that we can detect the mistakes

```

```python
df['Education Level']=df['Education Level'].str.title() #making every entries' format into title
df['Education Level']=df['Education Level'].str.strip() #removing the leading and trailing spaces of entries if there are any
df['Education Level']=df['Education Level'].str.replace('#','',regex=False) #removing the '#' character in the entries if there are any
df['Education Level']=df['Education Level'].str.replace('*','',regex=False) #removing '*' character in the entries if there are any.
df['Education Level']=df['Education Level'].fillna(df['Education Level'].mode()[0]) #filling the missing values of education level column with the first element of the modes  of the education level column.
df['Education Level'].unique() #checking the education level column values
```

```python
df['Employment Status'].value_counts() # checking the counts of the employment status types
```

```python
df['Employment Status']=df['Employment Status'].str.title() #making the format of each entry in employment status column as title
df['Employment Status']=df['Employment Status'].str.strip() #removing the leading and trailing space in the entries if there are any
df['Employment Status']=df['Employment Status'].str.replace('#','',regex=False) #removing the '#' character in the entries if there are any
df['Employment Status']=df['Employment Status'].str.replace('*','',regex=False) #removing the '*' character in the entries if there are any
df['Employment Status']=df['Employment Status'].fillna(df['Employment Status'].mode()[0]) # filling missing values of employment status column with the first element of the modes of the employment status column
df['Employment Status'].unique() #checking the employment status types
```

```python
df['Income'] = df['Income'].astype(str).str.replace(r'[\$,]', '', regex=True).astype(float) # firstly converting the type of Income column into string then later on we removed $  and ',' symbols lastly converted type into float
df['Income'] = df['Income'].fillna(df['Income'].median()) #filling the missing values of Income column with the median of the Income column
df['Income'] = df['Income'].astype(int) #converting the data type of Income column into integer
df['Income'].isnull().sum() #checking whether there are missing values Income column


```

```python
df['Stress Score']=df['Stress Score'].fillna(df['Stress Score'].median()) #filling missing values of Stress Score column with median of that column
df['Stress Score'].isnull().sum() #checking count of the missing values in stress score column
```

```python
df['Mental Health Score']=df['Mental Health Score'].fillna(df['Mental Health Score'].median())#filling missing values of Mental Health Score column with median of that column
df['Mental Health Score'].isnull().sum()#checking count of the missing values in mental health score column
```

```python
df['Drug Type'].value_counts() #checking counts of drug types so that we can detect the typos
```

```python
df['Drug Type']=df['Drug Type'].str.strip() #removing the leading and the trailing spaces in drug type column if there are any
df['Drug Type']=df['Drug Type'].str.title() #making the format of each entry in drug type column as title
df['Drug Type']=df['Drug Type'].str.replace('#','',regex=False) #removing the '#' character in drug type column
df['Drug Type']=df['Drug Type'].str.replace('*','',regex=False) #removing the '*' character in drug type column
df['Drug Type']=df['Drug Type'].fillna(df['Drug Type'].mode()[0]) #filling  the missing values of drug type column with the first element of modes of drug type column
df['Drug Type'].unique() #checking the types of drugs  in final form.
```

```python
df['Drug Use Frequency']=df['Drug Use Frequency'].fillna(df['Drug Use Frequency'].median()) #filling the missing values of drug use frequency column with the median of the drug use frequency column
df['Drug Use Frequency']=df['Drug Use Frequency'].astype(int) #converting the data type into integer
df['Drug Use Frequency'].isnull().sum() #checking the count of missing values in drug use frequency column
```

```python

df['Peer Influence']=df['Peer Influence'].fillna(df['Peer Influence'].mean()) #filling the missing values of peer influence column with mean of the peer influence column
df['Peer Influence'].isnull().sum() #checking the count of missing values in peer influence column
```

```python
df['Family History'].value_counts() #checking the counts of answer types in family history column
```

```python
df['Family History']=df['Family History'].str.strip() #removing the leading and trailing space in family history column
df['Family History']=df['Family History'].str.title() #making the format of each entry as title
df['Family History']=df['Family History'].str.replace('#','',regex=False) #removing the '#' character in family history column if there are any
df['Family History']=df['Family History'].str.replace('*','',regex=False) #removing the '*' character in family history column if there are any
df['Family History']=df['Family History'].fillna(df['Family History'].mode()[0]) #filling missing values in family history column with the first element of the modes of family history column
df['Family History'].unique() #checking the answer types of family history column
```

```python
df['Exercise Days']=df['Exercise Days'].fillna(df['Exercise Days'].median()) #filling the missing values of exercise days column with the median of that column
df['Exercise Days']=df['Exercise Days'].astype(int) #converting the dta type of the exercise days column into integer
df['Exercise Days'].isnull().sum() #checking the count of missing values in exercise days column

```

```python
df['Drug Awareness Program'].value_counts() #checking the answer types in drug awareness program column
```

```python
df['Drug Awareness Program']=df['Drug Awareness Program'].str.title() #making the format of each entry in drug awareness column as title
df['Drug Awareness Program']=df['Drug Awareness Program'].str.strip() #removing the leading and trailing space in each entry of drug awareness column if there are any
df['Drug Awareness Program']=df['Drug Awareness Program'].str.replace('#','',regex=False) #removing the '#' character in the entries of drug awareness column if there are any
df['Drug Awareness Program']=df['Drug Awareness Program'].str.replace('*','',regex=False) #removing the '*' character in the entries of drug awareness column if there are any
df['Drug Awareness Program']=df['Drug Awareness Program'].fillna(df['Drug Awareness Program'].mode()[0]) #filling the missing values of drug awareness column with first element of the modes of drug awareness program column
df['Drug Awareness Program'].unique() # checking the answer types of drug awareness column
```

```python
df['Risk Index']=df['Risk Index'].fillna(df['Risk Index'].mean()) #filling the missing values of risk index column with the mean of the risk index column
df['Risk Index'].isnull().sum() #checking the count of missing values
```
